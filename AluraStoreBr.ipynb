### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Calcula o total vendas para cada loja
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print("\n\033[1mTotal do faturamento de cada loja.\033[0m\n")
print(f"Total de faturamento da Loja 1: {faturamento_loja1}")
print(f"Total de faturamento da Loja 2: {faturamento_loja2}")
print(f"Total de faturamento da Loja 3: {faturamento_loja3}")
print(f"Total de faturamento da Loja 4: {faturamento_loja4}")

# Calcula o valor total do frete de cada loja
total_frete_loja1 = loja['Frete'].sum()
total_frete_loja2 = loja2['Frete'].sum()
total_frete_loja3 = loja3['Frete'].sum()
total_frete_loja4 = loja4['Frete'].sum()

print("\n\033[1mTotal do frete de cada loja.\033[0m\n")
print(f"Total do frete da Loja 1: {round(total_frete_loja1, 2)}")
print(f"Total do frete da Loja 2: {round(total_frete_loja2, 2)}")
print(f"Total do frete da Loja 3: {round(total_frete_loja3, 2)}")
print(f"Total do frete da Loja 4: {round(total_frete_loja4, 2)}")

# Calcula a média do frete de cada loja
media_frete_loja1 = loja['Frete'].mean()
media_frete_loja2 = loja2['Frete'].mean()
media_frete_loja3 = loja3['Frete'].mean()
media_frete_loja4 = loja4['Frete'].mean()

print("\n\033[1mMédia do frete de cada loja.\033[0m\n")
print(f"Média do frete da Loja 1: {round(media_frete_loja1, 2)}")
print(f"Média do frete da Loja 2: {round(media_frete_loja2, 2)}")
print(f"Média do frete da Loja 3: {round(media_frete_loja3, 2)}")
print(f"Média do frete da Loja 4: {round(media_frete_loja4, 2)}")

# Calcula a receita total de cada loja (usando o faturamento total - frete total
# para dar o resultado)

receita_loja1 = faturamento_loja1 - total_frete_loja1
receita_loja2 = faturamento_loja2 - total_frete_loja2
receita_loja3 = faturamento_loja3 - total_frete_loja3
receita_loja4 = faturamento_loja4 - total_frete_loja4

receitas = {
    "Loja 1": receita_loja1,
    "Loja 2": receita_loja2,
    "Loja 3": receita_loja3,
    "Loja 4": receita_loja4
}

maior_receita = max(receitas, key=receitas.get)
menor_receita = min(receitas, key=receitas.get)

print("\n\033[1mReceita de cada loja (faturamento - frete).\033[0m\n")
for loja, receita in receitas.items():
    print(f"Receita da {loja}: {round(receita, 2)}")
print("\n\033[1mLoja com maior e menor receita.\033[0m\n")
print(f"Loja com maior receita: {maior_receita} - R$ {round(receitas[maior_receita], 2)}")
print(f"Loja com menor receita: {menor_receita} - R$ {round(receitas[menor_receita], 2)}")


```

```python
import matplotlib.pyplot as plt

# Dados de receita (substitua pelos seus dados reais)
receitas = {
    "Loja 1": receita_loja1,
    "Loja 2": receita_loja2,
    "Loja 3": receita_loja3,
    "Loja 4": receita_loja4
}

lojas = list(receitas.keys())
valores = list(receitas.values())

# Criar o gráfico de barras
plt.figure(figsize=(8, 4))  # Ajustar o tamanho da figura
plt.bar(lojas, valores, color=['skyblue', 'lightcoral', 'lightgreen', 'lightpink'])  # Cores personalizadas
plt.xlabel("Lojas")
plt.ylabel("Receita Total")
plt.title("Receita Total por Loja")
plt.grid(axis='y', linestyle='--', alpha=0.7) # Adiciona uma grade no eixo y

# Adicionar os valores acima de cada barra
for i, v in enumerate(valores):
    plt.text(i, v, f'{v:.2f}', ha='center', va='bottom')


plt.show()
```

# 2. Vendas por Categoria


```python
# Agrupa as vendas por categoria em cada loja
vendas_por_categoria_loja1 = loja.groupby('Categoria do Produto')['Preço'].sum()
vendas_por_categoria_loja2 = loja2.groupby('Categoria do Produto')['Preço'].sum()
vendas_por_categoria_loja3 = loja3.groupby('Categoria do Produto')['Preço'].sum()
vendas_por_categoria_loja4 = loja4.groupby('Categoria do Produto')['Preço'].sum()

# Imprime os resultados
print("Vendas por categoria - Loja 1:")
print(vendas_por_categoria_loja1)
print("\nVendas por categoria - Loja 2:")
print(vendas_por_categoria_loja2)
print("\nVendas por categoria - Loja 3:")
print(vendas_por_categoria_loja3)
print("\nVendas por categoria - Loja 4:")
print(vendas_por_categoria_loja4)
```

```python
import matplotlib.pyplot as plt
# Criando os gráficos de pizza para cada loja
plt.figure(figsize=(10, 6))

plt.subplot(2, 2, 1)
plt.pie(vendas_por_categoria_loja1, labels=vendas_por_categoria_loja1.index, autopct='%1.1f%%', startangle=90)
plt.title('Loja 1')

plt.subplot(2, 2, 2)
plt.pie(vendas_por_categoria_loja2, labels=vendas_por_categoria_loja2.index, autopct='%1.1f%%', startangle=90)
plt.title('Loja 2')

plt.subplot(2, 2, 3)
plt.pie(vendas_por_categoria_loja3, labels=vendas_por_categoria_loja3.index, autopct='%1.1f%%', startangle=90)
plt.title('Loja 3')

plt.subplot(2, 2, 4)
plt.pie(vendas_por_categoria_loja4, labels=vendas_por_categoria_loja4.index, autopct='%1.1f%%', startangle=90)
plt.title('Loja 4')

plt.tight_layout()
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
def top_bottom_products(df, loja_name):
    vendas_produto = df.groupby('Produto')['Preço'].sum()
    mais_vendido = vendas_produto.idxmax()
    menos_vendido = vendas_produto.idxmin()
    print(f"Loja {loja_name}:")
    print(f"Produto mais vendido: {mais_vendido}")
    print(f"Produto menos vendido: {menos_vendido}")
    print("-" * 20)

top_bottom_products(loja, 1)
top_bottom_products(loja2, 2)
top_bottom_products(loja3, 3)
top_bottom_products(loja4, 4)
```

```python
import matplotlib.pyplot as plt

def plot_top_bottom_products(df, loja_name):
    vendas_produto = df.groupby('Produto')['Preço'].sum()
    total_vendas = vendas_produto.sum()

    mais_vendido = vendas_produto.idxmax()
    percentual_mais_vendido = (vendas_produto.max() / total_vendas) * 100

    menos_vendido = vendas_produto.idxmin()
    percentual_menos_vendido = (vendas_produto.min() / total_vendas) * 100

    produtos = [mais_vendido, menos_vendido]
    percentuais = [percentual_mais_vendido, percentual_menos_vendido]

    plt.figure(figsize=(6, 3))
    plt.plot(produtos, percentuais, marker='o')
    plt.xlabel("Produtos")
    plt.ylabel("Porcentagem de Vendas")
    plt.title(f"Produtos Mais e Menos Vendidos - Loja {loja_name}")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

plot_top_bottom_products(loja, 1)
plot_top_bottom_products(loja2, 2)
plot_top_bottom_products(loja3, 3)
plot_top_bottom_products(loja4, 4)
```

# 5. Frete Médio por Loja


```python
# Calcular o frete médio de cada loja
frete_medio_loja1 = loja['Frete'].mean()
frete_medio_loja2 = loja2['Frete'].mean()
frete_medio_loja3 = loja3['Frete'].mean()
frete_medio_loja4 = loja4['Frete'].mean()

# Criar um dicionário com os fretes médios
fretes_medios = {
    'Loja 1': frete_medio_loja1,
    'Loja 2': frete_medio_loja2,
    'Loja 3': frete_medio_loja3,
    'Loja 4': frete_medio_loja4
}

# Imprimir os resultados
print("\nFretes médios de cada loja:")
for loja, frete_medio in fretes_medios.items():
    print(f"{loja}: R${frete_medio:.2f}")
```

```python
import matplotlib.pyplot as plt

# Dados dos fretes médios (substitua pelos seus dados reais)
fretes_medios = {
    'Loja 1': frete_medio_loja1,
    'Loja 2': frete_medio_loja2,
    'Loja 3': frete_medio_loja3,
    'Loja 4': frete_medio_loja4
}

# Criar o gráfico de linhas
lojas = list(fretes_medios.keys())
valores_fretes = list(fretes_medios.values())

plt.figure(figsize=(8, 6))  # Ajuste o tamanho da figura conforme necessário
plt.plot(lojas, valores_fretes, marker='o', linestyle='-')

# Adicionar rótulos e título
plt.xlabel('Lojas')
plt.ylabel('Frete Médio (R$)')
plt.title('Comparação do Frete Médio por Loja')

# Adicionar valores acima de cada ponto no gráfico
for i, valor in enumerate(valores_fretes):
    plt.annotate(f'{valor:.2f}', (lojas[i], valor), textcoords="offset points", xytext=(0,10), ha='center')

# Mostrar o gráfico
plt.grid(True) # Adicionar grid ao gráfico para melhor visualização
plt.show()
```
